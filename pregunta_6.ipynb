# Pregunta 6
¿Qué directores se repiten más en el catálogo de cada plataforma?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfdisney_per = pd.read_csv('data/personas_disney.csv')
dfhulu_per = pd.read_csv('data/personas_hulu.csv')
dfmax_per = pd.read_csv('data/personas_max.csv')
dfnetflix_per = pd.read_csv('data/personas_netflix.csv')
dfparamount_per = pd.read_csv('data/personas_paramount.csv')
dfprime_per = pd.read_csv('data/personas_prime.csv')
```

Identificamos la columna que contenga el rol de cada persona en la producción de la película.

```python
dfdisney_per.head()
```

En este caso, es la columna "category". Ahora nos toca separar a todos los directores.

```python
dfdisney_dir = dfdisney_per[dfdisney_per['category'] == 'director'].reset_index(drop=True)
dfdisney_dir.head()
```

Ahora nos gustaría ver qué directores se repiten más. Además, como en el DataFrame no aparecen los nombres, usaremos un archivo .tsv que relaciona los valores de la columna "nconst" con sus respectivos nombres.

```python
imdb_name = pd.read_csv('data/name.basics.tsv', sep='\t')
imdb_name.head()
```

```python
ndisney_dir = dfdisney_dir['nconst'].value_counts().reset_index()
ndisney_dir.head()
```

```python
ndisney_dir = ndisney_dir.merge(imdb_name[['nconst', 'primaryName']])
ndisney_dir.head()
```

Repetimos el proceso con las demás plataformas.

```python
dflist = [dfhulu_per, dfmax_per, dfnetflix_per, dfparamount_per, dfprime_per]
dflist_dir = [ndisney_dir]

for plataforma in dflist:
    dfplataforma_dir = plataforma[plataforma['category'] == 'director'].reset_index(drop=True)
    nplataforma_dir = dfplataforma_dir['nconst'].value_counts().reset_index()
    nplataforma_dir = nplataforma_dir.merge(imdb_name[['nconst', 'primaryName']])
    dflist_dir.append(nplataforma_dir)
```

Como el primer dato de cada DataFrame contiene el director más repetido, podemos comparar a los directores de cada plataforma con un gráfico.

```python
colors = {
    'Disney': 'mediumturquoise',
    'Hulu': 'springgreen',
    'HBO Max': 'royalblue',
    'Netflix': 'firebrick',
    'Paramount': 'dodgerblue',
    'Prime Video': 'darkcyan'
}

plt.figure(figsize=12,6)

# falta agregar los 3 primeros elementos de cada DataFrame al gráfico

plt.xlabel('Directores')
plt.ylabel('Cantidad de veces que se repite')
plt.title('Los 3 Directores con más películas de cada plataforma')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
