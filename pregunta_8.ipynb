## Creacion dataframes

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

netflix = pd.read_csv('data/imdb_netflix.csv')
hulu = pd.read_csv('data/imdb_hulu.csv')
paramount = pd.read_csv('data/imdb_paramount.csv')
disney = pd.read_csv('data/imdb_disney.csv')
hbomax = pd.read_csv('data/imdb_max.csv')
prime = pd.read_csv('data/imdb_prime.csv')
```

```python
for i in [netflix, hulu, paramount, disney, hbomax, prime]:
    i['genres'] = i['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
netflix_genres = netflix.explode('genres')
hulu_genres = hulu.explode('genres')
paramount_genres = paramount.explode('genres')
disney_genres = disney.explode('genres')
hbomax_genres = hbomax.explode('genres')
prime_genres = prime.explode('genres')
```

```python
netflix_family = netflix_genres[netflix_genres['genres'] == 'Family']
hulu_family = hulu_genres[hulu_genres['genres'] == 'Family']
paramount_family = paramount_genres[paramount_genres['genres'] == 'Family']
disney_family = disney_genres[disney_genres['genres'] == 'Family']
hbomax_family = hbomax_genres[hbomax_genres['genres'] == 'Family']
prime_family = prime_genres[prime_genres['genres'] == 'Family']
nombre_df = {'Netflix': netflix_family,
            'Hulu': hulu_family,
            'Paramount': paramount_family,
            'Disney+': disney_family,
            'HBO Max': hbomax_family,
            'Prime': prime_family}
```

## Cantidad de peliculas familiares

```python
resultados1 = {}
for i in nombre_df.keys():
    resultados1[i] = len(nombre_df[i])
resultados1
```

```python
plt.figure(figsize=(8,5))
plt.bar(resultados1.keys(), resultados1.values(), color='skyblue', edgecolor='black')

plt.xlabel('Plataforma')
plt.ylabel('Cantidad de películas familiares')
plt.title('Cantidad de películas familiares por plataforma')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(resultados1.values()):
    plt.text(i, v + 1, str(v), ha='center', va='bottom', fontsize=9)

plt.show()
```

## Cantidad peliculas familiares con rating mayor a 8

```python
resultados2 = {}
for i in nombre_df.keys():
    resultados2[i] = len(nombre_df[i][nombre_df[i]['averageRating'] >= 8.0])
resultados2
```

```python
plt.figure(figsize=(8,5))
plt.bar(resultados2.keys(), resultados2.values(), color='skyblue', edgecolor='black')

plt.xlabel('Plataforma')
plt.ylabel('Cantidad de películas familiares')
plt.title('Cantidad peliculas familiares con rating mayor a 8')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(resultados2.values()):
    plt.text(i, v + 1, str(v), ha='center', va='bottom', fontsize=9)

plt.show()
```

```python
df_graf = pd.DataFrame({
    'Total': resultados1,
    'Rating ≥ 8': resultados2,
    'Rating ≥ 7': resultados3,
    'Rating ≥ 6': resultados4,
    'Rating ≥ 5': resultados5
})

df_graf = df_graf[['Total', 'Rating ≥ 8', 'Rating ≥ 7', 'Rating ≥ 6', 'Rating ≥ 5']]

plt.figure(figsize=(10,6))
for columna in df_graf.columns:
    plt.plot(df_graf.index, df_graf[columna], marker='o', label=columna)

plt.title('Cantidad de películas familiares por plataforma y rating')
plt.xlabel('Plataforma')
plt.ylabel('Cantidad de películas')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Filtro de rating')
plt.tight_layout()

plt.savefig('img/peliculas_familiares_por_rating.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
