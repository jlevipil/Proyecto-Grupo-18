### Primero hacemos los imports

```python
import pandas as pd 
import numpy as np
```

### Abrimos el DataFrame con toda la informaci√≥n

```python
df_max = pd.read_csv("api_max.csv")
df_max.head(10)
```

```python
imbd_basics =  pd.read_csv('title.basics.tsv', sep='\t')
imbd_basics.head()
```

```python
imbd_ratings = pd.read_csv("title.ratings.tsv", sep='\t')
imbd_ratings.head()
```

```python
imdb_crew = pd.read_csv('title.crew.tsv', sep='\t')
imdb_crew.head()
```

```python
imdb_name = pd.read_csv('name.basics.tsv', sep='\t')
imdb_name.head()
```

### Hacemos la creaci√≥n del DataFrame

```python
df_main1 = df_max.merge(imbd_basics, how='left', left_on='imdb_id', right_on='tconst')
df_main1.head()
```

```python
df_main2 = df_main1.merge(imbd_ratings, how='left', left_on='imdb_id', right_on='tconst')
df_main2.head()
```

```python
df_main3 = df_main2.merge(imdb_crew, how='left', left_on='imdb_id', right_on='tconst')
df_main3.head()
```

```python
df_main4 = df_main3[['id', 'imdb_id', 'primaryTitle', 'originalTitle', 'titleType', 'startYear', 
                    'isAdult', 'runtimeMinutes', 'genres', 'averageRating', 'numVotes', 'directors', 'writers']]
df_main4
```

```python
#Ocupe chatgpt para hacer la uni√≥n de los codigos del imb con los nombres de los directores, escritores y , el prompt que ocupe fue guiame para poder hacer la un√≠on de los codigos imb con los nombres de los actores, directores, escritores de los siguientes archivos (name.basics.tsv, title.principals.tsv, df_main4)
#Primero se crea un mapa de ids el cual contiene los nombres
mapa_nombres = dict(zip(imdb_name['nconst'], imdb_name['primaryName']))
#Luego cree una funci√≥n para pasar de los codigos a los nombres, el cual su unica funcion es recibir los codigos y el mapa de nombre y va haciendo la busqueda una por uno, y si alg√∫n id no esta en el diccionario lo omite.
def convertir_codigos_a_nombres(codigos, df_map):
    if pd.isna(codigos):
        return None
    lista_codigos = codigos.split(',')
    nombres = []
    for cod in lista_codigos:
        cod = cod.strip()
        if cod in df_map:
            nombres.append(df_map[cod])
    return ', '.join(nombres)
#Luego a√±adimos las nuevas columnas con los nombres de los directores y escritores
df_main4['directors_names'] = [convertir_codigos_a_nombres(x, mapa_nombres) for x in df_main4['directors']]

df_main4['writers_names'] = [convertir_codigos_a_nombres(x, mapa_nombres) for x in df_main4['writers']]
```

```python
imdb_principals = pd.read_csv('title.principals.tsv', sep='\t', dtype=str)

# Filtramos solo actores y actrices
df_actors = imdb_principals[imdb_principals['category'].isin(['actor', 'actress'])]

# Agrupamos y concatenamos todos los actores por tconst
df_actors_grouped = (
    df_actors
    .groupby('tconst', as_index=False)
    .agg({'nconst': ', '.join})
    .rename(columns={'nconst': 'actors'})
)



```

```python
df_main4 = df_main4.merge(df_actors_grouped, left_on='imdb_id', right_on='tconst', how='left')

# Crear una nueva columna con los nombres de los actores
df_main4['actors_names'] = df_main4['actors'].apply(
    lambda x: convertir_codigos_a_nombres(x, mapa_nombres)
)


```

```python
#Guardamos el csv
df_main4.to_csv("imbd_max.csv", index=False)
```
