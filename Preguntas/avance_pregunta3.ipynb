```python
import pandas as pd
import matplotlib.pyplot as plt

# Lista de DataFrames y nombres
plataformas = {
    'Netflix': netflix,
    'Hulu': hulu,
    'Paramount': paramount,
    'Disney': disney,
    'HBO Max': hbomax,
    'Prime Video': prime
}

# Rango completo de aos para todos los DataFrames
aos_min = min(df['year'].min() for df in plataformas.values())
aos_max = max(df['year'].max() for df in plataformas.values())
aos_completos = range(aos_min, aos_max+1)

# Preparar conteos por ao (llenando con 0 donde no hay ttulos)
conteos = {}
for nombre, df in plataformas.items():
    conteo = df['year'].value_counts().reindex(aos_completos, fill_value=0).sort_index()
    conteos[nombre] = conteo

# Grfico de lneas
plt.figure(figsize=(20,6))
for nombre, conteo in conteos.items():
    plt.plot(conteo.index, conteo.values, marker='o', linestyle='-', label=nombre)

plt.xticks(range(aos_min, aos_max+1, 1), rotation=90)  # todos los aos en eje X
plt.xlabel('Ao')
plt.ylabel('Cantidad de ttulos')
plt.title('Cantidad de ttulos por ao por plataforma')
plt.legend()
plt.grid(True)
plt.show()
```
