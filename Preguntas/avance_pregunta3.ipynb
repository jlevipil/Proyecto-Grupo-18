```python
import pandas as pd
import matplotlib.pyplot as plt

netflix = pd.read_csv('data/imdb_netflix.csv')
hulu = pd.read_csv('data/imdb_hulu.csv')
paramount = pd.read_csv('data/imdb_paramount.csv')
disney = pd.read_csv('data/imdb_disney.csv')
hbomax = pd.read_csv('data/imdb_max.csv')
prime = pd.read_csv('data/imdb_prime.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lista de DataFrames y nombres
plataformas = {
    'Netflix': netflix,
    'Hulu': hulu,
    'Paramount': paramount,
    'Disney': disney,
    'HBO Max': hbomax,
    'Prime Video': prime
}

# Rango completo de años para todos los DataFrames
años_min = min(df['year'].min() for df in plataformas.values())
años_max = max(df['year'].max() for df in plataformas.values())
años_completos = range(años_min, años_max+1)

# Preparar conteos por año (llenando con 0 donde no hay títulos)
conteos = {}
for nombre, df in plataformas.items():
    conteo = df['year'].value_counts().reindex(años_completos, fill_value=0).sort_index()
    conteos[nombre] = conteo

# Gráfico de líneas
plt.figure(figsize=(20,6))
for nombre, conteo in conteos.items():
    plt.plot(conteo.index, conteo.values, marker='o', linestyle='-', label=nombre)

plt.xticks(range(años_min, años_max+1, 1), rotation=90)  # todos los años en eje X
plt.xlabel('Año')
plt.ylabel('Cantidad de títulos')
plt.title('Cantidad de títulos por año por plataforma')
plt.legend()
plt.grid(True)
plt.show()
```
