# **Pregunta 5**

**¿Qué actores se repiten más en el catálogo de cada plataforma?**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Leemos los csv de **"personas_(plataforma).csv"**

```python
df_netflix5 = pd.read_csv("data/personas_netflix.csv")
df_disney5 = pd.read_csv("data/personas_disney.csv")
df_hulu5 = pd.read_csv("data/personas_hulu.csv")
df_max5 = pd.read_csv("data/personas_max.csv")
df_paramount5 = pd.read_csv("data/personas_paramount.csv")
df_prime5 = pd.read_csv("data/personas_prime.csv")
```

Leemos el archivo **name-basics.tsv** en el cual se relaciona el nconst y el nombre del actor

```python
df_names = pd.read_csv("tsv/name.basics.tsv", sep='\t')
```

Filtramos solo los actores y actrices

```python
df_actores_netflix = df_netflix5[df_netflix5['category'].isin(['actor', 'actress'])]
df_actores_disney = df_disney5[df_disney5['category'].isin(['actor', 'actress'])]
df_actores_hulu = df_hulu5[df_hulu5['category'].isin(['actor', 'actress'])]
df_actores_max = df_max5[df_max5['category'].isin(['actor', 'actress'])]
df_actores_paramount = df_paramount5[df_paramount5['category'].isin(['actor', 'actress'])]
df_actores_prime = df_prime5[df_prime5['category'].isin(['actor', 'actress'])]
```

Unimos los dos DataFrame por **nconst** 

```python
df_name_netflix = df_actores_netflix.merge(df_names[['nconst', 'primaryName']], on='nconst', how='left')
df_name_disney = df_actores_disney.merge(df_names[['nconst', 'primaryName']], on='nconst', how='left')
df_name_hulu = df_actores_hulu.merge(df_names[['nconst', 'primaryName']], on='nconst', how='left')
df_name_max = df_actores_max.merge(df_names[['nconst', 'primaryName']], on='nconst', how='left')
df_name_paramount = df_actores_paramount.merge(df_names[['nconst', 'primaryName']], on='nconst', how='left')
df_name_prime = df_actores_prime.merge(df_names[['nconst', 'primaryName']], on='nconst', how='left')
```

Contamos cuantas veces aparece cada actor

```python
repetidos_netflix = df_name_netflix['primaryName'].value_counts().reset_index()
repetidos_disney = df_name_disney['primaryName'].value_counts().reset_index()
repetidos_hulu = df_name_hulu['primaryName'].value_counts().reset_index()
repetidos_max = df_name_max['primaryName'].value_counts().reset_index()
repetidos_paramount = df_name_paramount['primaryName'].value_counts().reset_index()
repetidos_prime = df_name_prime['primaryName'].value_counts().reset_index()
```

```python
repetidos_netflix.columns = ['Actor/Actriz', 'Cantidad_apariciones']
repetidos_disney.columns = ['Actor/Actriz', 'Cantidad_apariciones']
repetidos_hulu.columns = ['Actor/Actriz', 'Cantidad_apariciones']
repetidos_max.columns = ['Actor/Actriz', 'Cantidad_apariciones']
repetidos_paramount.columns = ['Actor/Actriz', 'Cantidad_apariciones']
repetidos_prime.columns = ['Actor/Actriz', 'Cantidad_apariciones']
```

### **Función para graficar los actores que más se repiten**

```python
def graficar_top_actores(df, plataforma, color):
    top = df.head(10)
    plt.figure(figsize=(8,5))
    plt.barh(top['Actor/Actriz'], top['Cantidad_apariciones'], color=color)
    plt.gca().invert_yaxis()  # de mayor a menor de arriba a abajo
    plt.title(f"Top 10 actores más repetidos en {plataforma}", fontsize=14)
    plt.xlabel("Cantidad de apariciones")
    plt.ylabel("Actor/Actriz")
    plt.grid(axis='x', linestyle='--', alpha=0.6)
    plt.show()
```

### **Netflix**

```python
graficar_top_actores(repetidos_netflix, "Netflix", "red")
```

- Actor que se repite más veces en el catalogo de Netflix es: **Shah Rukh Khan**

### **Disney**

```python
graficar_top_actores(repetidos_disney, "Disney", "blue")
```

- Actor que se repite más veces en el catalogo de Disney es: **Jim Cummings**

### **Hulu**

```python
graficar_top_actores(repetidos_hulu, "Hulu", "green")
```

- Actor que se repite más veces en el catalogo de Hulu es: **Nicolas Cage**

### **Max**

```python
graficar_top_actores(repetidos_max, "Max", "purple")
```

- Actor que se repite más veces en el catalogo de Max es: **Grey DeLisle**

### **Paramount**

```python
graficar_top_actores(repetidos_paramount, "Paramount", "gray")
```

- Actor que se repite más veces en el catalogo de Paramount es: **Tom Cruise**

### **Prime**

```python
graficar_top_actores(repetidos_prime, "Prime", "lightblue")
```

- Actor que se repite más veces en el catalogo de Prime es: **Michael Paré**
