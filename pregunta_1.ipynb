## Pregunta 1

¿Qué plataforma presenta mayor cantidad de películas?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Abrimos el api de las plataformas que estaremos trabajando
df_disney = pd.read_csv("data/imdb_disney.csv")
df_hulu = pd.read_csv("data/imdb_hulu.csv")
df_max = pd.read_csv("data/imdb_max.csv")
df_netflix = pd.read_csv("data/imdb_netflix.csv")
df_prime = pd.read_csv("data/imdb_prime.csv")
df_paramount = pd.read_csv("data/imdb_paramount.csv")
```

```python
#Como los datas ya se encuentran limpios no es necesario hacer ningún tipo de limpieza
total_peliculas_disney = len(df_disney)
total_peliculas_hulu = len(df_hulu)
total_peliculas_max = len(df_max)
total_peliculas_netflix = len(df_netflix)
total_peliculas_prime = len(df_prime)
total_peliculas_paramount = len(df_paramount)
print(f"Disney cuenta con {total_peliculas_disney} películas en su catálogo")
print(f"Hulu cuenta con {total_peliculas_hulu} películas en su catálogo")
print(f"Max cuenta con {total_peliculas_max} películas en su catálogo")
print(f"Netflix cuenta con {total_peliculas_netflix} películas en su catálogo")
print(f"Prime cuenta con {total_peliculas_prime} películas en su catálogo")
print(f"Paramount cuenta con {total_peliculas_paramount} películas en su catálogo")
```

```python
#Graficamos para ver mejor la comparación, primero guardamos un dataframe con los datos respectivos
data = {"Plataforma": ["Disney","Hulu", "Max", "Netflix", "Prime","Paramount"],
        "Cantidad Películas": [total_peliculas_disney, total_peliculas_hulu, total_peliculas_max, total_peliculas_netflix, total_peliculas_prime, total_peliculas_paramount]}
df = pd.DataFrame(data)
#Elegi una gama de colores para las barras de la plataforma
colores = ["#F94144", "#F3722C", "#F8961E","#F9C74F", "#F9844A", "#F6AE2D"]  
#Aplique un fondo visual para que se viera más bonito
sns.set_style("whitegrid")
plt.figure(figsize=(10,8))
sns.barplot(data=data,palette=colores, x="Plataforma", y="Cantidad Películas", hue = "Plataforma")
plt.title("Cantidad de películas por plataformas", fontsize=16)
#También agregue lineas -- para el estilo del grafico
plt.grid(axis="y", linestyle = "--", alpha=0.6)
```

Por lo tanto la plataforma que tiene una mayor cantidad de películas en su catálogo es Prime Video con un total de 7387
