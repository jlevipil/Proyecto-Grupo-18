### Pregunta 4

¿Cúal es la plataforma es la que concentra la mayor cantidad de películas con mejores calificaciones?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cargamos los datasframes del imbd
df_netflix = pd.read_csv("data/imdb_netflix.csv")
df_hulu = pd.read_csv("data/imdb_netflix.csv")
df_disney = pd.read_csv("data/imdb_disney.csv")
df_max = pd.read_csv("data/imdb_max.csv")
df_prime = pd.read_csv("data/imdb_prime.csv")
df_paramount = pd.read_csv("data/imdb_paramount.csv")
```

```python
#Ordenamos los ratings del imdb de mayor a menor calificaciones
top_netflix = df_netflix[df_netflix["averageRating"] >= 8.5].copy()
top_hulu = df_hulu[df_hulu["averageRating"] >= 8.5].copy()
top_disney = df_disney[df_disney["averageRating"] >= 8.5].copy()
top_max = df_max[df_max["averageRating"] >= 8.5].copy()
top_prime = df_prime[df_prime["averageRating"] >= 8.5].copy()
top_paramount = df_paramount[df_paramount["averageRating"] >= 8.5].copy()
#Quiero unirlos y crear un solo dataframe para poder las calificaciones top que en esta las definiremos >= 8.5
#Por lo que creare una columna donde se vea el nombre de la plataforma
top_netflix["Plataforma"] = "Netflix"
top_hulu["Plataforma"] = "Hulu"
top_disney["Plataforma"] = "Disney"
top_max["Plataforma"] = "Max"
top_prime["Plataforma"] = "Prime"
top_paramount["Plataforma"] = "Paramount"
#Ahora utilizamos concat para dejarlos en un solo dataframe
top_plataformas = pd.concat([top_netflix, top_hulu, top_disney, top_max, top_prime, top_paramount], ignore_index=True)
```

```python
#Ordenamos el top de mayor a menor
top_plataformas = top_plataformas.sort_values(by="averageRating",ascending=False)
top_plataformas.head()
```

```python
cantidad_netflix = len(top_plataformas[top_plataformas["Plataforma"] == "Netflix"])
cantidad_hulu = len(top_plataformas[top_plataformas["Plataforma"] == "Hulu"])
cantidad_prime = len(top_plataformas[top_plataformas["Plataforma"] == "Prine"])
cantidad_max = len(top_plataformas[top_plataformas["Plataforma"] == "Max"])
cantidad_disney = len(top_plataformas[top_plataformas["Plataforma"] == "Disney"])
cantidad_paramount = len(top_plataformas[top_plataformas["Plataforma"] == "Paramount"])
print(f"Netflix tiene {cantidad_netflix} de películas mejores calificadas")
print(f"Hulu tiene {cantidad_hulu} de películas mejores calificadas")
print(f"Max tiene {cantidad_max} de películas mejores calificadas")
print(f"Prime tiene {cantidad_prime} de películas mejores calificadas")
print(f"Disney tiene {cantidad_disney} de películas mejores calificadas")
print(f"Paramount tiene {cantidad_paramount} de películas mejores calificadas")
```

```python
#Elijimos una paleta de colores para el histograma
colores = {"Disney": "#A3A9D1","Hulu": "#D4A5C2","Max": "#C0C0C0","Netflix": "#9FB7B9","Prime": "#7A9BAE","Paramount": "#DB8BF2"}
#Definimos un orden en caso de que cualquier cosa
orden = ["Disney", "Hulu", "Max", "Netflix", "Prime", "Paramount"]
#Graficamos
plt.figure(figsize=(10,6))

ax = sns.histplot(data=top_plataformas,x="averageRating",hue="Plataforma",hue_order=orden,multiple="dodge",shrink=0.8,palette=colores,edgecolor="black",alpha=0.4)



plt.xlim(8.5, 10)
plt.title("Distribución de calificaciones ≥ 8.5 por plataforma", fontweight="bold")
plt.xlabel("Rating promedio (IMDB)")
plt.ylabel("Número de películas")
plt.xticks(np.arange(8.5, 10.1, 0.1))
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()


```

Por lo tanto max concentra la mayor cantidad de películas mejores rankeadas
